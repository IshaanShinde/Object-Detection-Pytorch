```python
import cv2
from matplotlib import pyplot as plt
```

# Testing Capture Devices

```python
cap = cv2.VideoCapture(0)
```

```python
# get frame from capture device
ret, frame = cap.read()
```

```python
plt.imshow(frame)
```

```python
plt.imshow(cv2.color(frame, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# release capture device
cap.release()
```

```python
def getFrame():
    cap = cv2.VideoCapture(0)
    ret, frame = cap.read()
    cv2.imwrite('frameCapture.jpg', frame)
    cap.release()
```

```python
getFrame()
```

# Real Time Rendering

```python

cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()


    cv2.imshow('videoCapture', frame)

    # if q is pressed it closes the capture
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
